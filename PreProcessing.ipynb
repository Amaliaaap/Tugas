**1. Persiapan Data**


```python
# Mengimport library yang digunakan (Pandas)
import pandas as pd
```

```python
# membaca file CSV dan menyimpan datanya ke dalam sebuah DataFrame.
df = pd.read_csv('/content/drive/MyDrive/Data Mining/movie_sample_dataset.csv')
df.head()
```

**2. Memeriksa Data**



```python
# Menampilkan baris pertama dari dataset
df.head(1)
```

```python
# Memeriksa informasi umum tentang dataset
df.info()
```

```python
import numpy as np

# Memeriksa jumlah missing value di setiap kolom
print(df.isnull().sum())
```

**3. Membersihkan Data**



```python
df.shape
```

```python
df.loc[:, 'color'] = df['color'].str.strip().str.capitalize()
df.loc[:, 'country'] = df['country'].str.strip().str.upper()
df = df[(df['duration']>=0)&
              (df['gross']>=0)&
              (df['title_year']>=0)&
              (df['budget']>=0)&
              (df['imdb_score']>=0)&
              (df['movie_facebook_likes']>=0)]
df = df.replace('N/A', pd.NA)
df.dropna(inplace=True)
df
```

```python
# Mengecek apakah masih ada Missing data
print(df.isnull().sum())
```

4. Transformasi Data

```python
# Mengecek apa tipe data awal
df.dtypes
```

```python
# Mengubah tipe data
df['gross'] = df['gross'].astype('int')
df['budget'] = df['budget'].astype('int')
df.dtypes
```

```python
# Memisahkan kolom genre
print(df['genres'].head())
```

```python
# Memisahkan genre
genres_split = df['genres'].str.split('|', expand=True)
print(genres_split.head())
```

```python
# Membuat kolom baru
genres_split.columns = [f'genre_{i+1}' for i in range(genres_split.shape[1])]

# Gabungkan dengan DataFrame asli
df = pd.concat([df, genres_split], axis=1)
```

Kita memberikan nama pada kolom baru dengan format genre_1, genre_2, dan seterusnya, berdasarkan jumlah kolom hasil pemisahan. Kita menggunakan pd.concat([df, genres_split], axis=1) untuk menggabungkan DataFrame asli dengan DataFrame genre yang baru dibuat secara horizontal.


```python
df.drop(columns=['genres'], inplace=True)

# dengan df.drop(columns=['genres'], inplace=True), kita menghapus kolom 'genres' yang sudah tidak diperlukan lagi setelah pemisahan.

# tampilkan dataframe
df.head()
```
